```python
import numpy as  np
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('data/cohort_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```python
df.info()
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
df['transaction_month'] = df['transaction_date'].apply(get_month)
```

```python
group = df.groupby('customer_id')['transaction_month']
```

```python
df['cohort_month'] = group.transform('min')
```

```python
df.head()
```

```python
def get_month_int(dframe, column):
    year = dframe[column].dt.year
    month = dframe[column].dt.month
    day = dframe[column].dt.day
    return year, month, day
```

```python
invoice_year, invoice_month, _ = get_month_int(df, 'transaction_month')
```

```python
cohort_year, cohort_month, _ = get_month_int(df, 'cohort_month')
```

```python
year_diff = invoice_year - cohort_year
month_diff = invoice_month - cohort_month
```

```python
df['index_cohort'] = year_diff * 12 + month_diff
```

```python
df.head()
```

```python
grouping = df.groupby(['cohort_month', 'index_cohort'])
cohort_data = grouping['customer_id'].apply(pd.Series.nunique)
```

```python
cohort_data = cohort_data.reset_index()
```

```python
cohort_data.info()
```

```python
cohort_data['cohort_month'] = pd.DatetimeIndex(cohort_data['cohort_month']).to_period('M') # tanda
```

```python
cohort_count = cohort_data.pivot(columns='index_cohort', index='cohort_month', values='customer_id')
```

```python
cohort_count
```

```python
retention = cohort_count.divide(cohort_count[0], axis=0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
plt.title('Cohort Analysis for Customer Retention')
sns.heatmap(retention.round(3), annot=True, fmt='.0%')
plt.yticks(rotation=0)
plt.show()
```

```python
retention.drop([0], axis=1, inplace=True)
retention.insert(loc=0, column='user', value=cohort_count[0])
retention.round(2)
```
